#Opening a csv file from an URL

```python
import pandas as pd
import requests
from io import StringIO

url = 'https://example.com/data.csv'

# Download the CSV
response = requests.get(url)
csv_content = StringIO(response.text)

# Read into DataFrame
df = pd.read_csv(csv_content)

print(df.head())

```
