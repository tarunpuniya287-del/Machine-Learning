```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,noise=80,random_state=13)
plt.scatter(x,y)
```

```python
# Let's apply Ordinary least square (OLS)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='red')
```

```python
# let's apply Gradient Descent assuming slope in constant , m=78.35
# and let's assume the starting value for intercept b =0
y_pred=((78.35 * x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='blue',label='b = 0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0

loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
# let's take learning rate = 0.1
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
# calculate the new intercept
b=b-step_size
b
```

```python
y_pred1=((78.35 * x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='green',label='b = {}'.format(b))
plt.plot(x,y_pred,color='blue',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((78.35 * x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='red',label='OLS')
plt.plot(x,y_pred2,color='black',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='green',label='b = {}'.format(b))
plt.plot(x,y_pred,color='blue',label='b = 0')
plt.legend()
plt.show()
```

```python
# iteration 3
loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3=((78.35 * x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='red',label='OLS')
plt.plot(x,y_pred3,color='yellow',label='b = {}'.format(b))
plt.plot(x,y_pred2,color='black',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='green',label='b = {}'.format(b))
plt.plot(x,y_pred,color='blue',label='b = 0')
plt.legend()
plt.show()
```
