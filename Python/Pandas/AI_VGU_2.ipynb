```python
# Create a function to calculate factorial
# Factorial: 5! = 5*4*3*2*1
def factorial(n):
    result = 1
    for i in range(1,n+1):
        result = result * i
    return result

print(factorial(5))
```

```python
# Use list comprehension to flatten a 2D list

matrix = [[1,2],[3,4],[5,6]]
flat = [num for row in matrix for num in row]
print(flat)
```

```python
# Find the second largest number in a list

nums = [10,40,20,55]
first = second = float('-inf')

for n in nums:
    if n > first:
        second = first
        first = n
    elif n > second and n!= first:
        second = n
print(second)

# Concepts: Logic, comparison
```

```python
# Check if a number is prime
n = int(input('Enter a number:'))
is_prime = True

for i in range(2,int(n**0.5)+1): # This range is to determine all the factors of n
    if n%i == 0:
        is_prime = False
        break
print(is_prime)

# Concepts: Optimization, math
```

```python
# Removing duplicates and maintaining order
nums= [1,2,2,3,1,4]
unique = []
for i in nums:
    if i not in unique:
        unique.append(i)

print(unique)

# Concepts: Lists, conditions, loop
```

```python
# Find common elements between two lists
a = [1,2,3,4]
b = [3,4,5]

common = list(set(a) & set(b))
print(common)

# Concepts: Sets
```

```python
# Lambda function with map()
nums = [1,2,3,4]
squares = list(map(lambda x:x**2, nums))
print(squares)

# Concepts: Functional Programming
```
