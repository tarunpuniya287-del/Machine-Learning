#### K-NEAREST NEIGHBOURS (KNN)

```python
# What is KNN.?
# KNN or K-nearest neighbours is a supervised learning algorithm used mainly for classification.
# Sometimes for regression
# Simple idea: 'Tell me who your neighbours are and it will tell you who you are'


## Why do we use KNN.?
# We use KNN when:
# Data is small to medium
# Patterns are simple
# We want a non-mathematical model
# KNN dont train a complex model
# Its often used as:
# 1. A baseline model
# 2. A teaching tool
# 3. A quick solution for simple classification

## Key idea behind KNN
# KNN does not learn a formula
# Instead:
# 1. It stores all training data
# 2. When a new point comes:
#     1. Find the K closest data points
#     2. Takes a vote
#     3. Assigns the most common class


## What does K mean.?
# # K = Number of neighbours to consider
# Examples:
# K = 1 -> looks at only the nearest neighbour
# K = 3 -> looks at 3 nearest neighbours
# K = 5 -> looks at 5 nearest neighbours
# Usually odd values are chosen to avoid ties


# How does KNN find 'nearest'.?
# Using distant metrics, most commonly
# Euciledian distance = sqrt(square(x2-x1) + square(y2-y1))
# Measures how far two points are from each other

# Example:
# Imagine students classified as Pass(1) or Fail(0)
# Based on Study Hours
# If a new student studies for 5 hours, KNN:
# 1. Look at nearby students
# 2. Checks their results
# 3. Assign the majority class
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Imports the KNN Algorithm
```

```python
data = {
    'study_hours': [1,2,3,6,7,8],
    'result' : [0,0,0,1,1,1]
}

df = pd.DataFrame(data)
df.head()
```

```python
# X -> Features (study hours)
# Y -> Labels (0 = fail, 1= pass)
X = df[['study_hours']]
y = df['result']
```

```python
model = KNeighborsClassifier(n_neighbors = 3)
# Only uses 3 nearest neighbors
```

```python
model.fit(X,y)
# Stores the data
```

```python
model.predict([[5]])
# Finds 3 closest values to 5
# Takes majority vote
# Output is the majority class
```

```python
# Advantages of KNN
# 1. Very easy to understand
# 2. No training time
# 3. Works well for small datasets

# Disadvantages of KNN
# 1. Very slow for large datasets
# 2. Sensitive to noise (outliers, null values, duplicate values,spaces)
# 3. Needs feature scaling
# 4. Memory heavy
```

#### DECISION TREES

```python
# What is a decision tree.?
# A decision tree is also a supervised learning algorithm
# Uses if else rules
# Splits data into branches
# Make decision step by step
# Looks like a flowchart


## Why do we use Decision Trees.?
# 1. They are highly interpretable
# 2. Mimics human decision making
# 3. Works for both classfication and regression
# 4. Handle non-linear data very well

# How Does a decision tree work.?
# 1. Starts with the entire dataset
# 2. Finds the best feature to split
# 3. Splits data into branches
# 4. Repeats untill:
#  a. Pure Classes
#  b. Max depth reached

# Important Terms
# 1. Node : A decision point (e.g: 'is_age > 30')
# 2. Root Node: Topmost Decision
# 3. Leaf Node: Final Prediction (Output)

# How does it choose the best split.?
# Using metrics like:
# 1. Gini impurity
    # Gini impurity tells us how mixed(impure) the classes are at a node.
    # Gini impurity helps us answer which split or which node has the least mixed data
    # 0: perfectly pure node (only one class), 0.5: maximum impurity, higher value: more mixed data
    # Gini = 1 - [square(p(case)) + square(p(x)]
# 2. Entropy (Information Gain)
# Reduce Randomness and increase purity
```

```python
from sklearn.tree import DecisionTreeClassifier
# imports Decision Tree algorithm
```

```python
model = DecisionTreeClassifier()
model.fit(X,y)
# Learns Decision rules
```

```python
model.predict([[5]])
# Passes data through rules
# Returns class
```

```python
# Advantages of Decision Tree
# 1. Easy to visualize
# 2. No feature scaling needed
# 3. Handles categorical data

# Disadvantages
# 1. Unstable(small change -> Big tree)
# 2. Overfitting
```

#### RANDOM FOREST

```python
# What is random forest.?
# Random forest is an ensemble learning algorithm
# It takes many decision trees and takes a majority vote

# 'WISDOM OF THE CROWD'

# Overfitting: Model is too complex model
# Underfitting: Model is too simple
# Best fit: Best model

## Why do we need random forests.?
# Because single deciion tree often overfit.
# Random forest will reduce overfitting, improves accuracy, and improves stability.

## How does random forest work.?
# Create multiple datasets
# Train decision tree on each dataset
# Each tree will make a prediction
# Final result = Majority Vote

# Ensemble learning: Combination of multiple models
```

```python
from sklearn.ensemble import RandomForestClassifier
# Import Random forest algorithm
```

```python
model = RandomForestClassifier(n_estimators = 100)
# n_estimators are the number of decision trees
```

```python
model.fit(X,y)
model.predict([[10]])
```

```python
# # Advantages
# 1. Highly accurate
# 2. Handles large datasets
# 3. Less overfitting

# Disadvantages
# 1. Slower
# 2. Less interpretable
```

#### K-MEANS CLUSTERING (UNSUPERVISED LEARNING)

```python
# WHAT IS K-MEANS.?
# It is an unsupervised learning algorithm
# Used for clustering
# Groups data without labels

# Why do we use K-means.?
# Used when: 
# 1. No output labels is available
# 2. Find patterns
# 3. Grouping/Segementation

# Example:
# 1. Image compression
# 2. Customer segmentation
# 3. Market analysis

# What does K means in Kmeans.?
# K = number of clusters
# Example: K=2:-> two groups
# K=3 -> Three groups

# How does KMeans work.?
# 1. Chooses K cluster centers randomly
# 2. Assigns each point to nearest clusters
# 3. Recalculates the cluster centers
# 4. Repeat untill convergence

# Important terms
# Centroid: Mean point of cluster
```

```python
from sklearn.cluster import KMeans
# Imports Kmeans algorithm
```

```python
X = [[200],[220],[250],[1000],[1100],[1200]]
# Customer Spending Data
```

```python
kmeans = KMeans(n_clusters=2,random_state=42)
kmeans.fit(X)
```

```python
kmeans.labels_
# Shows cluster assignment
```

```python
plt.figure(figsize=[3,2])
plt.scatter(range(len(X)),X,c=kmeans.labels_)
plt.show()
```

```python
# Advantages
# 1. Simple and Fast
# 2. Works well for spherical clusters

# Disadvantages
# 1. Need to choose K manually
# 2. Sensitive to outliers
```
