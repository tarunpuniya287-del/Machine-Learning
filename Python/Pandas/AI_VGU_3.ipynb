#### NumPy

```python
## Numpy means numerical python.
## Numpy is used for fast numerical operations, arrays, matrices, and mathematical functions.
## Array is like a numerical storing lists. Arrays are much faster than lists and they use less space than lists.

import numpy as np # used to import numpy library

# Creation of numpy array
arr = np.array([10,20,30,40,50])
print(arr)

# Array properties
print(arr.shape) # Shape in numpy determines the number of elements in an array
print(arr.ndim) # Shows dimensions of array, number of dimensions
print(arr.dtype)  # Shows the datatype of array


## Mathematical Operations
print(arr + 5) # Adds 5 in every element of the array
print(arr * 2) # Multiplies every element with 2
print('The mean of this array is:',np.mean(arr)) # np.mean will take out the mean of the array
print('The median of this array is:',np.median(arr)) # np.median will take out the median of the array
print('The standard deviation of this array is:',np.std(arr)) # np.std will take out the standard deviation of the array.

## 2-D Array
matrix = np.array([[1,2,3],[4,5,6]])
print(matrix)
print('Column wise summation is:',matrix.sum(axis = 0)) # Column-wise addition of arrays
print('Row wise sumamtion is:',matrix.sum(axis = 1)) # Row-wise addition of arrays
```

#### PANDAS

```python
## This is used for data handling and data analysis
## Used for tabular data, used for data cleaning, used for data analysis, used for data manipulation.

import pandas as pd # Used for importing pandas library

## Creating Series and Dataframes
# Series: Its just like a column in an excel sheet
s = pd.Series([90,85,80]) # Single column is known as series
print(s)

# DataFrame: It is like the entire excelsheet
data = {
    'Name': ['Piyush','Devansh','Sachin'],
    'Marks': [85,86,87],
    'Passed': [True, True, True]
}
df = pd.DataFrame(data)
df

## Basic Dataframe Oeprations
df.head() # Shows first 5 rows
df.tail() # Shows the last 5 rows
df.info() # Shows entire inforamtion about the dataset
df.describe() # Shows descriptive stats (mean, median, Quartiles, valuecount, standarad deviation, max, min)


## Column and Row selection
df['Marks'] # Selects the particualr column
print(df[['Name','Marks']]) # Selects multiple columns
df.loc[0] #Selects by row index
df.iloc[1] # Selcts particular values


## Filtering Data
df[df['Marks'] > 85] # The outer df is to show all the filtered value which satisfied the condition

## Adding New column 
df['Grade'] = ['A','A','A'] # To add a new column, just write the name of the column with output values
df
```

#### MATPLOT AND SEABORN

```python
# Used to create basic plots and charts - Matplot
import matplotlib.pyplot as plt

## Line plot
x = [1,2,3,4]
y = [10,20,25,30]

plt.figure(figsize=[3,2]) # Used to adjust the size of the graph
plt.plot(x,y) # Plot is used for line plot
plt.xlabel('SMALL NUMBERS') # Writes the x axis label
plt.ylabel('BIG NUMBERS') # Writes the y-axis label
plt.title('LINE PLOT EXAMPLE') # Writes the title of the graph
plt.show() # Show is used to show the chart


## Bar Chart
names = ['A','B','C']
values = [20,35,30]

plt.figure(figsize=[3,2])
plt.bar(names,values) # .bar is used to create bar chart
plt.xlabel('Names')
plt.ylabel('Values')
plt.title('BAR CHART EXAMPLE')
plt.show()

## Histogram
data = np.random.normal(70,10,100) # Random generates ramdom values in this case 100 random values, 70 is mean and 10 is std dev
plt.figure(figsize=[3,2])
plt.hist(data) # .hist is used to plot histogram
plt.xlabel('Ranges')
plt.ylabel('Vaue Count')
plt.title('Histogram Example')
plt.show()
```

```python
## Seaborn
## Used for data visualzation
## Used for beautiful and statistical plots
## There are inbuilt datasets which you can import

# Improting seaborn
import seaborn as sns
## Load Sample dataset
df = sns.load_dataset('tips') # Loading a inbuilt dataset
print(df.head(6))

## Counplot
plt.figure(figsize=[3,2])
sns.countplot(x = 'sex', data=df) # Countplot is used to count numerical values for a categorical column
plt.title('Countplot of gender')
plt.show()

## Boxplot
plt.figure(figsize=[3,2])
sns.boxplot(x = 'day',y='total_bill',data=df) # Used to detect outliers
plt.show()

## ScatterPlot
plt.figure(figsize=[3,2])
sns.scatterplot(x='total_bill',y='tip',data=df) # Used to show collinearity between two numerical variables
plt.show()
```
