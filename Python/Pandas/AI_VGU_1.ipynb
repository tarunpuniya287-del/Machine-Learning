### OPERATORS

```python
a = 10
b = 3
# Arithematic Operators
print(a+b) # Adds
print(a-b) # Subtracts
print(a*b) # Multiplies
print(a/b) # Divides
print(a%b) # Modulus: Used to find remainders
print(a//b) # Floor division: Used to find quotients

# Assignment Operators


# Comparison Operators
print(a>b)
print(a==b)
```

### Control Structures

```python
# if-else statements
age = 20
if age >= 18:
    print('Eligible to vote')
else:
    print('N/A')
```

```python
#if-elif-else statements
marks = 55
if marks >= 90:
    print('A')
elif marks >= 60:
    print('B')
else:
    print('C')
```

```python
# Loops
# for loop: Iterates over a sequence
for i in range(5):
    print(i)
```

```python
# and while loop: Iterates untill the condition is true
count = -1
while count <= 3:
    print(count)
    count = count + 1
```

```python
# Print a student report card
# Take 3 inputs of marks
# Calculate the total marks and print it
# Calculate the percentage and print it
# If marks are greater than 250 print A
# If marks are greater than 150 print B
# Else C
marks1 = int(input('Enter marks of your first subject:'))
marks2 = int(input('Enter marks of your second subject:'))
marks3 = int(input('Enter marks of your third subject:'))
print('--------------- Report Card --------------------')
total = marks1 + marks2 + marks3
print('Total Marks of the subjects:',total)
percentage = total / 3
print('Percentage acquired:',percentage)
if total > 250:
    print('GRADE A')
elif total > 150:
    print('GRADE B')
else:
    print('GRADE C')
# --------------- Report Card --------------------
# marks 1 = 25
# marks 2 = 75
# marks 3 = 70
# Total aggregate marks = 170
# Percentage = 56.64
# Grade B
```
